The code below is to generate a numeric presentation of the face image

```python
# Function to encode image to a NumPy array and then to a base64 string
def image_to_base64(image_path):
    # Read the image using OpenCV
    img = cv2.imread(image_path)
    
    # Convert the image to a NumPy array
    img_array = np.array(img)
    
    # Convert the NumPy array to binary format
    _, img_encoded = cv2.imencode('.jpg', img)
    img_binary = img_encoded.tobytes()
    
    # Convert binary to base64 string for storage
    img_base64 = base64.b64encode(img_binary).decode('utf-8')
    
    return img_base64

# Function to decode base64 to NumPy array and display the image
def base64_to_image(base64_str):
    # Decode base64 to binary data
    img_binary = base64.b64decode(base64_str)
    
    # Convert the binary data back to a NumPy array
    nparr = np.frombuffer(img_binary, np.uint8)
    img = cv2.imdecode(nparr, cv2.IMREAD_COLOR)
    
    # Display the image using OpenCV
    cv2.imshow('Decoded Image', img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

Opening the json and getting the name of the face , so those which is already process can be skip , because we want to add new faces only

```python
with open("./embeddings.json" , "r") as f:
    embeddings_data = json.load(f)
```

```python
names = []
for n in embeddings_data:
    names.append(n)

print(names)
```

```python
for folder in os.listdir("./sample_data"):
    if folder not in names:
        if os.path.isdir(os.path.join("./sample_data" , folder)):
            for image in os.listdir(os.path.join("./sample_data" , folder)):
                if image.endswith(".jpg"):
                    #temp = image_to_base64(os.path.join("./sample_data" , folder , image))
                    embeddings = DeepFace.represent(os.path.join("./sample_data" , folder , image), model_name = models[6] , detector_backend="retinaface" , align=True)
                    lst_face_embeddings.append(embeddings[0]["embedding"])
                    #lst_images.append(temp)
            embeddings_data[folder] = {"embeddings" : lst_face_embeddings}
            lst_face_embeddings = []
        
```
